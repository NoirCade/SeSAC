### 1. 모듈은 파이썬을 통해 무언가 원하는 작업을 하기 위해 필요한 도구!  
이미지를 보정하거나 회전 시키는 등의 작업을 도와주는 PIL이나 OpenCV, 수치화된 데이터들을 조작하게 해주는 numpy, pandas 등

모듈의 설치 및 로드 방법
1)	!pip install 명령어 사용하여 모듈 설치
2)	import 명령어 이용해서 해당 모듈을 로드하여 사용
3)	모듈의 위치 지정해줘야 할 경우 from ~ import ~

모듈을 사용하면 전체적인 코드 구조가 훨씬 단순해지고 필요한 기능을 쉽게 활용할 수 있다!



### 2.	출력은 print(), input() 값은 늘 String, 변수는 값을 저장하는 공간!
print() 함수를 통해서 다양한 연산을 거친 값을 출력하여 확인할 수 있다.  
```print(‘{}~{}’.format(a,b)) ```
```print(‘%d, %d’%(a, b)) ```  
와 같이 **format**을 이용하거나 %를 사용하여 출력 형식을 조정할 수 있다.

input() 함수를 이용하여 변수에 저장한 값은 항상 **String type**이 된다.  
따라서 해당 값을 변수에 저장하여 사용할 때에는 늘 유의하여야 한다.  
만약 input을 시도할 때에 사용자의 실수가 발생할 여지가 있을 경우, **try/except** 함수를 이용하여 예외 처리를 하는 것이 좋다.

변수의 이름을 지정할 때에 **예약어(키워드)**를 사용할 수는 있지만,  
그럴 경우 기존 예약어에 저장된 함수를 사용할 수 없게 되므로, 사용하지 않도록 한다.  
또한 숫자로 시작하거나 공백을 포함할 수 없으며 _(언더 바)를 제외한 모든 특수문자를 사용할 수 없다!

### 3. PILLOW 모듈의 사용

파이썬에는 이미지 처리 모듈이 크게 3가지 정도 있다
1. PILLOW (가장 쉬움, 이미지 파일로 open)
2. OpenCV (파이썬, c등 여러 언어에서 사용, 숫자로 된 array로 open)
    - 비디오 처리, 포토샵 등에서 사용되는 필터, 눈/코/입 검출 등 기술에 이용
3. sk-img (이미지 타입 변형/이미지 전처리 특화)
