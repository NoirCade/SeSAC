인공지능이 학습할 때 지도학습(input하는 x값, 결과값 y값을 같이 제공)  
컴퓨터 비전은: 지도학습 
 
x값: 이미지 1장, y값: 그 이미지에 해당하는 라벨  
    (ex: 0 or 1 or 2 or 숫자값, 사용자가 0이 '개'인지, 1이 '고양이'인지 알아서 정리해둬야함)  

그래서 인공지능 학습 시에 수없이 많은 **kwargs들이 들어간다!

```python
import matplotlib.pyplot as plt

def ImgListShow(XData, yData):
    for x,y in zip(XData,yData):
        plt.imshow(x)
        plt.title(y)
        plt.show()
```

```python
def myArgsImgShow(**kwargs):
    x,y=kwargs['xDataList'],kwargs['yDataList']
    for x,y in zip(x,y):
        plt.imshow(x)
        plt.title(y)
        plt.show()
```
