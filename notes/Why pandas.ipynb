각종 데이터를 정리하고 다루는데 도움이 되는 판다스!

막상 써보면 엑셀이랑 큰 차이 없어보이는데 굳이 왜 판다스?

1. 데이터 량의 차이
    적은 양의 데이터만 가지고 정리하고 정렬하여 사용할 때는 엑셀도 좋지  
    그러니까 일반 직군에서는 엑셀을 많이 사용하고, 단순한 차트를 만들 때에는 좋다  
    그렇지만 수십만 수백만의 자료값들을 이용해야 할 때는? 통계 분석에 특화된 차트를 써야할 때는?  
    -> 빅데이터를 다루려면 결국 판다스를 사용할 수밖에 없다
    반대로 대량의 데이터/통계분석 특화/데이터 자동화 세가지 경우가 아니라면 굳이 파이썬을 써야할 이유도 없다는 의미!

2. 시각화의 차이
    앞서 말한대로 단순한 차트는 제공하지만 통계분석을 위한 추가 모듈을 쓰려면 반드시 파이썬을 거쳐야한다
