```python
def fileOpen(fileList):
    xData,yData=[],[]
    for file in fileList:
        filename=file.split('\\')[-1].split('.')[0]
        img=Image.open(file).resize((200,300))
        img=np.array(img)
        xData.append(img)
        yData.append(filename)
    
    return xData,yData
```

```python
fileList=glob('C:\\Users\\bluecom015\\Desktop\\SeSAC\\data/img/*.jpg')
imgData=fileOpen(fileList)
# print(imgData)

def imgPltShow(num=0,**kwargs):
    #print(kwargs)
    for xD,yD in zip(kwargs['x'],kwargs['y']):
        plt.imshow(xD+num)
        plt.title(yD)
        plt.show()


img,label=fileOpen(fileList)
imgPltShow(-50,x=img,y=label)
```
