```python
folder=r'C:\Users\bluecom015\Desktop\SeSAC\practice\day3\img'

from glob import glob
from PIL import Image

fileList=glob(folder + '\\*')
imgList=[]

for x in fileList:
    img=Image.open(x).resize((100,200))
    imgArr=np.array(img)
    imgList.append(imgArr)

imgList=np.array(imgList)
imgList.shape

plt.imshow(imgList[0])
```

```python
for cnt,i in enumerate(imgList):
    plt.subplot(1,len(imgList),cnt+1)
    plt.imshow(i)
```
