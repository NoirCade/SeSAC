````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

speech_1 = "Ladies and gentlemen, we stand at a critical juncture in the history of our planet. The evidence of climate change is undeniable, and the consequences of our actions are becoming increasingly severe. It is time for us to take bold and decisive action to address the urgent threat of climate change."

speech_2 = "As global citizens, we must recognize that the effects of climate change are not limited by borders. Rising temperatures, extreme weather events, and melting ice caps affect us all. We have a collective responsibility to safeguard our planet for current and future generations."

speech_3 = "In the face of this crisis, we cannot afford to be complacent. We must transition to a sustainable and low-carbon economy, embracing renewable energy sources and adopting eco-friendly practices. Our choices today will shape the world our children and grandchildren inherit."

speech_4 = "The time for debate is over; now is the time for action. Governments, businesses, and individuals must collaborate to reduce greenhouse gas emissions, protect endangered ecosystems, and promote environmental sustainability. Together, we can build a resilient and sustainable future."

speech_5 = "Climate change is not just an environmental issue; it is also a matter of social justice. The most vulnerable communities are disproportionately affected by the impacts of climate change. It is our moral duty to ensure that our response is inclusive and that no one is left behind."

speeches = [speech_1, speech_2, speech_3, speech_4, speech_5]

plt.figure(figsize=(15, 10))

for cnt, speech in enumerate(speeches):
    plt.subplot(2, 3, cnt+1)
    
    wordcloud = WordCloud(width=400, height=200, background_color="white").generate(speech)
    
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.title(f'Speech {cnt+1}')
    plt.axis("off")

plt.tight_layout()
plt.show()
````

````python
문자열리스트=['강 나닫 강아지 고양이 개 강아지 고양이', '가나', 'ㄷ나다']
txt=문자열리스트[0]
wordList=txt.split(' ')

dataList=[]
for word in wordList:
    wordLen=len(word)
    if wordLen>=3:
        print(word, '-->', len(word))
        dataList.append(word)

dataList
````
