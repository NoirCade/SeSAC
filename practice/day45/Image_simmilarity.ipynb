```python
from glob import glob

img_root = 'C:/Users/bluecom015/Desktop/SeSAC/practice/day44/extracted_images/Images - 1/Images - 1'
img_list = glob(img_root + '/*.jpg')
img_list
```

```python
import cv2

hist_list = []

for img in img_list:
    np_img = cv2.imread(img)
    hsv_img = cv2.cvtColor(np_img, cv2.COLOR_BGR2HSV)
    hist = cv2.calcHist([hsv_img], [0, 1], None, [180, 256], [0, 180, 0, 256])
    cv2.normalize(hist, hist, 0, 1, cv2.NORM_MINMAX)
    hist_list.append(hist)

hist_list
```

```python
train_labels = glob('C:/Users/bluecom015/Downloads/resized_woods/train/labels/*.txt')
boxes = []

for label in train_labels:
    box = []
    with open(label, 'r') as f:
        for line in f:
            box.append(line.strip())
    boxes.append(box)

boxes
```

```python
no_box = []
for idx, box in enumerate(boxes):
    if box == []:
        no_box.append(idx)

len(no_box)
```

```python
test_img = img_root + '/100000052.jpg'
test_img
```

```python
np_test = cv2.imread(test_img)
hsv_test = cv2.cvtColor(np_test, cv2.COLOR_BGR2HSV)
hist_test = cv2.calcHist([hsv_test], [0, 1], None, [180, 256], [0, 180, 0, 256])
cv2.normalize(hist_test, hist_test, 0, 1, cv2.NORM_MINMAX)

sim_list = []

for hist in hist_list:
    similarity = cv2.compareHist(hist_test, hist, cv2.HISTCMP_CORREL)
    sim_list.append(similarity)

sim_list
```

```python
good_woods = []
for idx, sim in enumerate(sim_list):
    if sim>=0.93:
        good_woods.append(idx)

len(good_woods)
```
