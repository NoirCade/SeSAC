```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
import os
```

```python
pathFolder = "../day47/train/spaceship/"
os.makedirs(pathFolder,exist_ok=True)
xTrainName = "XTrain.pkl"
yTrainName = "yTrain.pkl"

with open(pathFolder+xTrainName,'rb') as f1:
    X = pickle.load(f1)
```

```python
column1_data = X[:, 0]
column2_data = X[:, 1]

# 히스토그램 그리기
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(column1_data, bins=30, color='blue', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Column 1')

plt.subplot(1, 2, 2)
plt.hist(column2_data, bins=30, color='blue', alpha=0.7)
plt.xlabel('Column 2')
plt.ylabel('Frequency')
plt.title('Histogram of Column 2')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pickle

# 넘파이 어레이 데이터 가져오기
pathFolder = "../day47/train/spaceship/"
xTrainName = "XTrain.pkl"
with open(pathFolder+xTrainName,'rb') as f1:
    X = pickle.load(f1)

# 컬럼별 히스토그램 그리기
num_cols = X.shape[1]  # 데이터의 컬럼 수
fig, axs = plt.subplots(1, num_cols, figsize=(15, 5))  # 한 줄에 컬럼 개수만큼의 플롯 생성

for i in range(num_cols):
    axs[i].hist(X[:, i], bins=30, color='blue', alpha=0.7)
    axs[i].set_xlabel(f'X{i+1}')
    axs[i].set_ylabel('Frequency')
    axs[i].set_title(f'Distribution of X{i+1}')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pickle

# 넘파이 어레이 데이터 가져오기
pathFolder = "../day47/train/spaceship/"
xTrainName = "XTrain.pkl"
with open(pathFolder+xTrainName,'rb') as f1:
    X = pickle.load(f1)

X = np.array((X*2) -1)

# 컬럼별 히스토그램 그리기
num_cols = X.shape[1]  # 데이터의 컬럼 수
fig, axs = plt.subplots(1, num_cols, figsize=(15, 5))  # 한 줄에 컬럼 개수만큼의 플롯 생성

for i in range(num_cols):
    axs[i].hist(X[:, i], bins=30, color='blue', alpha=0.7)
    axs[i].set_xlabel(f'X{i+1}')
    axs[i].set_ylabel('Frequency')
    axs[i].set_title(f'Distribution of X{i+1}')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
