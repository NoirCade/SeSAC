```python
import pandas as pd
import os

path = 'C:/Users/kth53/Downloads/earthquakes.csv'

data = pd.read_csv(path)
data
```

```python
data.info()
```

```python
data_drop = data[['time', 'latitude', 'longitude', 'depth', 'mag']]
```

```python
data_drop = data_drop.dropna()
```

```python
data_drop.info()
```

```python
data_drop.head()
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(data_drop)

# Í∞Å Ïª¨ÎüºÏùò Í∞íÎì§Ïùò Î∂ÑÌè¨Î•º ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú ÎÇòÌÉÄÎÇ¥Í∏∞
fig, axs = plt.subplots(1, len(df.columns[1:]), figsize=(15, 5))

for i, col in enumerate(df.columns[1:]):
    axs[i].hist(df[col], bins=10, color='blue', alpha=0.7)
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Frequency')
    axs[i].set_title(f'Distribution of {col}')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
