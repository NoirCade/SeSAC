```python
import seaborn as sns
import matplotlib.pyplot as plt

# 남성인 승객만 추출
male_passengers = titanic[titanic['Sex'] == 'male']

# 나이를 3개의 구간으로 나누기
age_bins = [0, 18, 35, 80]
age_labels = ['0-18', '19-35', '36-80']
male_passengers['AgeGroup'] = pd.cut(male_passengers['Age'], bins=age_bins, labels=age_labels, right=False)

# 시각화
plt.figure(figsize=(10, 6))
sns.countplot(x='AgeGroup', hue='Survived', data=male_passengers)
plt.title('Survival by Age Group for Male Passengers')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
female_passengers = titanic[titanic['Sex'] == 'female']
female_passengers['AgeGroup'] = pd.cut(female_passengers['Age'], bins=age_bins, labels=age_labels, right=False)

# 시각화
plt.figure(figsize=(10, 6))
sns.countplot(x='AgeGroup', hue='Survived', data=female_passengers)
plt.title('Survival by Age Group for Female Passengers')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(15, 6))

# 남성 데이터 시각화
plt.subplot(1, 2, 1)
sns.countplot(x='AgeGroup', hue='Survived', data=male_passengers)
plt.title('Survival by Age Group for Male Passengers')
plt.xlabel('Age Group')
plt.ylabel('Count')

# 여성 데이터 시각화
plt.subplot(1, 2, 2)
sns.countplot(x='AgeGroup', hue='Survived', data=female_passengers)
plt.title('Survival by Age Group for Female Passengers')
plt.xlabel('Age Group')
plt.ylabel('Count')

plt.tight_layout()  # 서브플롯 간 간격 조절
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 남성이면서 사망한 승객만 추출
dead_male_passengers = titanic[(titanic['Sex'] == 'male') & (titanic['Survived'] == 0)]
survived_male_passengers = titanic[(titanic['Sex'] == 'male') & (titanic['Survived'] == 1)]

# 나이를 3개의 구간으로 나누기
age_bins = [0, 18, 35, 80]
age_labels = ['0-18', '19-35', '36-80']
dead_male_passengers['AgeGroup'] = pd.cut(dead_male_passengers['Age'], bins=age_bins, labels=age_labels, right=False)
survived_male_passengers['AgeGroup'] = pd.cut(survived_male_passengers['Age'], bins=age_bins, labels=age_labels, right=False)

# 시각화
plt.figure(figsize=(12, 6))

# 나이 및 승객 등급에 따른 사망한 남성 데이터 시각화
plt.subplot(1, 2, 1)
sns.countplot(x='AgeGroup', hue='Pclass', data=dead_male_passengers)
plt.title('Dead Male Passengers: Age Group and Pclass')
plt.xlabel('Age Group')
plt.ylabel('Count')

# 나이 및 승객 등급에 따른 생존한 남성 데이터 시각화
plt.subplot(1, 2, 2)
sns.countplot(x='AgeGroup', hue='Pclass', data=survived_male_passengers)
plt.title('Survived Male Passengers: Age Group and Pclass')
plt.xlabel('Age Group')
plt.ylabel('Count')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 남성 데이터 추출
male_passengers = titanic[titanic['Sex'] == 'male']

# 나이를 3개의 구간으로 나누기
age_bins = [0, 18, 35, 80]
age_labels = ['0-18', '19-35', '36-80']
male_passengers['AgeGroup'] = pd.cut(male_passengers['Age'], bins=age_bins, labels=age_labels, right=False)

# 시각화
plt.figure(figsize=(12, 6))

# 연령 구간 및 객실 등급에 따른 남성 생존/사망 비율 시각화
sns.barplot(x='AgeGroup', y='Survived', hue='Pclass', data=male_passengers, ci=None)
plt.title('Survival Ratio of Male Passengers by Age Group and Pclass')
plt.xlabel('Age Group')
plt.ylabel('Survival Ratio')

plt.show()
```
