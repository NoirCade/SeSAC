```python
import os
import shutil

# 폴더 경로
folder_path = 'C:\\Users\\user\\Desktop\\프로젝트\\TL_2.trajectory_후기-궤적'
image_path='C:\\Users\\user\\Desktop\\프로젝트\\TS_1.motion_f-인사이드_성인 - 복사본'
# 대상 폴더 경로 (이미지 파일을 복사할 폴더)
destination_folder_path = 'C:\\Users\\user\\Desktop\\프로젝트\\Copied_Images'

# 이미지 파일 확장자
image_extension = '.jpg'

# JSON 파일 확장자
json_extension = '.json'

# 이미지 파일 목록 가져오기
image_files = [file for file in os.listdir(image_path) if file.endswith(image_extension)]

# JSON 파일 목록 가져오기
json_files = [file for file in os.listdir(folder_path) if file.endswith(json_extension)]

# 이미지 파일 중에서 JSON 파일과 이름이 일치하는 파일만 복사하기
for image_file in image_files:
    # 이미지 파일 이름에서 '_a' 전의 부분 추출
    image_name_prefix = image_file.split('_a')[0]

    # 대상 JSON 파일 이름 생성
    target_json_name = f"{image_name_prefix}_b.json"

    # 해당 JSON 파일이 있는지 확인
    matching_json = target_json_name in json_files
    

    # 일치하는 JSON 파일이 있으면 이미지 파일을 대상 폴더로 복사
    if matching_json:
        source_path = os.path.join(image_path, image_file)
        destination_path = os.path.join(destination_folder_path, image_file)

        try:
            shutil.copy(source_path, destination_path)
            print(f"Copied: {image_file}")
        except PermissionError as e:
            print(f"Permission error while copying {image_file}: {e}")

print(image_name_prefix)
print(matching_json)
```
