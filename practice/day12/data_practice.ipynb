```python
data['성별'].fillna(data['성별'].mode()[0], inplace=True)
data.info()
```

```python
print(data['영업점명'].value_counts())
print(data['식음메뉴명(한글)'].value_counts().sort_values(ascending=False))
print(data['성별'].value_counts())
print(data['고객여권발급국가명'].value_counts())
```

```python
print(data['주문수량'].groupby(data['영업점명']).sum())
print(data['주문수량'].groupby(data['식음메뉴명(한글)']).sum().sort_values(ascending=False))
print(data['주문수량'].groupby(data['성별']).sum())
print(data['주문수량'].groupby(data['고객여권발급국가명']).sum().sort_values(ascending=False))
```

```python
print(data['주문수량'].groupby([data['영업점명'],data['고객여권발급국가명']]).sum().sort_values(ascending=False))
```
