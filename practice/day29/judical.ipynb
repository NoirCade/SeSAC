```python
df['판례내용'][0]

import re

text = df['판례내용'][0]

# 정규 표현식을 사용하여 '【이    유】' 이후의 내용을 추출
result = re.search(r'【이\s+유】\s*([^\n]*)', text)
if result:
    extracted_text = result.group(1)
    print(extracted_text)
```

```python
numList = []
for num, x in enumerate(df['판례내용']):
    result = re.search(r'【이\s+유】\s*([^\n]*)', x)
    if result:
        extracted_text = result.group(1)
    cnt=len(extracted_text)
    if cnt<2:
        numList.append(num)

numList, len(numList)
```

```python
df['이유']=df['판례내용'].str.split('【이    유】').str[1]
df.head()
```

```python
df.info()
```

```python
df_re = df[['판례내용', '이유']].dropna()
df_re
```

```python
pos_list = []

okt = Okt()
for r in df_re['이유']:
    ps = okt.pos(r)
    for p in ps:
        if (p[1]=='Noun') or (p[1]=='Adjective'):
            pos_list.append(p)

pos_list
```

판례내용에서 '이유' 이후만 별도의 자료로 만들고  
명사와 형용사만 토큰화 시킨 후 문자열(str)로 제작하여 tf-idf에 넣을 자료를 작성한 후  
csv로 저장하여 다시 불러온다
