```python
from zipfile import ZipFile

with ZipFile('./spaceship-titanic.zip', 'r') as z:
    z.extractall('sst_csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./sst_csv/train.csv')
data.info()
```

```python
data.head()
```

```python
(data['Transported'] == True).sum(), (data['Transported'] == False).sum()
```

```python
data = data.drop(columns = ['HomePlanet', 'Cabin', 'Destination', 'Name'])
```

```python
drop_data = data.dropna()
(drop_data['Transported'] == True).sum(), (drop_data['Transported'] == False).sum()
```

```python
drop_data['ServiceCost'] = drop_data['RoomService'] + drop_data['FoodCourt'] + drop_data['ShoppingMall'] + drop_data['Spa'] + drop_data['VRDeck']
drop_data
```

```python
processed_data = drop_data.drop(columns = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Transported'])
processed_data
```

```python
one_hot_Cryo = processed_data['CryoSleep'].astype(int)
one_hot_VIP = processed_data['VIP'].astype(int)
one_hot_Cryo, one_hot_VIP
```

```python
processed_data = processed_data.drop(columns=['CryoSleep', 'VIP'])
processed_data['CryoInt'] = one_hot_Cryo
processed_data['VIPInt'] = one_hot_VIP
processed_data
```

```python
processed_label = drop_data[['PassengerId', 'Transported']]
processed_label['Transported'] = processed_label['Transported'].astype(int)
processed_label
```

```python
processed_data.to_csv('./processed_train.csv', index=False)
processed_label.to_csv('./processed_label.csv', index=False)
```

```python
import numpy as np
import pickle

final_data = processed_data.drop(columns='PassengerId')
pkl_data = np.array(final_data)

with open('XTrain.pkl', 'wb') as f:
    pickle.dump(pkl_data, f)

final_label = processed_label.drop(columns='PassengerId')
pkl_label = np.array(final_label)

with open('yTrain.pkl', 'wb') as f:
    pickle.dump(pkl_label, f)
```
