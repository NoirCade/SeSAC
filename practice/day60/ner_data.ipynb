```python
with open('train_data.txt', 'r') as f:
    data = f.readlines()

data
```

```python
new_data = []

for dat in data:
    new_dat = []
    if dat != '\n':
        dat = dat.split('\t')

        if (dat[1] != '．') and (dat[1] != '.'):
            new_dat.append(dat[1])
            new_dat.append(dat[2].split('\n')[0])
            
            if dat[2].split('\n')[0] == 'PER_B':
                new_dat.append(1)
            else:
                new_dat.append(0)

            new_data.append(new_dat)

new_data
```

```python
import pandas as pd
columns = ['String', 'label', 'person']

d = pd.DataFrame(new_data, columns=columns)

d
```

```python
import pickle

with open('ner.pickle', 'wb') as f:
    pickle.dump(d, f)
```
