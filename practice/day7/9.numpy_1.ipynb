```python
from PIL import Image
from glob import glob

imgList=glob(r'C:\Users\bluecom015\Desktop\SeSAC\data\img\*.jpg')
img=Image.open(imgList[0])
img
```

```python
import numpy as np
img=np.array(img)

## array type은 변수.shape가 가능
## list type는 np.shape(변수)로 shape 확인 가능
img.shape, np.shape(img)    # 행(가로) 308, 열(세로) 408, 3채널(RGB) 형태의 이미지
```

```python
# tmp=np.reshape(img, (1,308,408,3))  # img.reshape((1,308,408,3))과 동일 (이미 numpy array 상태이므로)
tmp=img[np.newaxis,...]
tmp.shape
```
